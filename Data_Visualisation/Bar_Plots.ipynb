```python
import numpy as np
```

```python
y = np.random.randint(1, 100, 5)
```

```python
x = np.arange(5)
```

```python
import matplotlib.pyplot as plt
```

```python
y1 = np.random.randint(1, 100, 5)
x1 = np.arange(5)
```

```python
plt.figure(figsize = (5,6))
plt.bar(x, y, width = 0.4,align = 'edge', label = 'India', tick_label = ['A', 'B', 'C', 'D', 'E'])
for i in range(len(x)):
    plt.text(x[i], y[i], str(y[i]))

plt.bar(x1, y1, width = -0.4, align = 'edge', label = 'USA')
plt.legend()

plt.xlabel('teriitory')
plt.ylabel('milk production')
plt.title('Territory wise milk production')
plt.show()
```
