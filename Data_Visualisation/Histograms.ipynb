```python
np.random.seed(42)
Xsn = np.random.randn(100000)
sigma = 8
u = 70
X = np.round(Xsn*sigma + u)

print(X)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8')
```

```python
plt.hist(X, alpha = 0.5) # So alpha decides the transparency of the image
plt.ylabel('Frequency Count')
plt.xlabel('Marks')
for i in range(len(X)):
    plt.text(X[i], str(X[i]))
plt.show()
```
