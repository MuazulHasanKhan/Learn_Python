```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Line Plot

x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
```

```python
plt.plot(x, y1, color = 'red', label = 'Apple')
plt.plot(x, y2, color = 'green',label = 'orange')
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Prices of fruit over time')
plt.legend()
plt.show()
```
