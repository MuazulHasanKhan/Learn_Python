```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
sns.barplot(x = 'sex' , y = 'total_bill', data = tips ) #
```

```python
import numpy as np
```

```python
sns.barplot(x = 'sex' , y = 'total_bill', data = tips, estimator = np.std ) #we have added an estimator by default it is mean 
```

```python
 sns.countplot(x = 'sex', data = tips)#aggregation function is always count plot
```

```python
sns.countplot(x = 'time', data = tips)
```

```python
# Boxplot helps us to find out outliers in the data
sns.boxplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.boxplot(x = 'day', y = 'total_bill', data = tips, hue = 'smoker')
# The line is the median
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips)
```

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips, hue = 'sex', split = True)
```

```python
# Distribution plot: it allows us to depict the distribution of a variable
```

```python
sns.distplot(tips['total_bill'], bins = 20, kde = False) #the line is called kernel density estimation
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
 #jointplot combines the distribution of two variable
```

```python
sns.jointplot(x ='total_bill' , y = 'tip' , data = tips)
```

```python
sns.jointplot(x ='total_bill' , y = 'tip' , data = tips, kind = 'scatter')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue = 'sex')
```

```python
tips.head()
```

```python
tips_corr = tips.corr(numeric_only = True)
```

```python
#heat map should have variables on both the rows and columns

tips_corr
```

```python
sns.heatmap(tips_corr, annot = True)
```
